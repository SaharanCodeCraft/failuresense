# Descriptive Statistical Analysis

This section performs descriptive statistical analysis to summarize and understand the central tendency, spread, and relationships among numerical features in the dataset.


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"D:\FailureSense_MLProj\failuresense\data\raw\ai4i2020.csv")
```

```python
numeric_features = df.select_dtypes(include="number")
numeric_features.head()
```

```python
numeric_features.describe()
```

The `describe()` function provides:
- Mean
- Standard deviation
- Minimum and maximum values
- 25th, 50th (median), and 75th percentiles


```python
numeric_features.median()
```

```python
numeric_features.median()
```

```python
numeric_features.sum()
```

```python
numeric_features.var()
```

```python
numeric_features.std()
```

```python
numeric_features.quantile([0.1, 0.25, 0.5, 0.75, 0.9])
```

Correlation Matrix

```python
correlation_matrix = numeric_features.corr()
correlation_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

Covariance Matrix

```python
covariance_matrix = numeric_features.cov()
covariance_matrix
```

## Statistical Observations

- Sensor features show varying ranges and variability
- Certain temperature-related features exhibit positive correlation
- Differences in scale across features justify the need for feature scaling
- Statistical summaries confirm the absence of abnormal values
