## What is Generative AI?

Generative AI refers to models capable of generating new content such as text, images, or code based on learned patterns from data.  
Examples include large language models (LLMs) used for explanation, summarization, and decision support.

In predictive maintenance systems, Generative AI is commonly used to:
- Explain model predictions
- Generate maintenance recommendations
- Assist human operators in decision-making


Simulated failure prediction input

```python
failure_case = {
    "Air temperature [K]": 304,
    "Process temperature [K]": 313,
    "Rotational speed [rpm]": 2500,
    "Torque [Nm]": 55,
    "Tool wear [min]": 210,
    "Predicted failure probability": 0.82
}
```

Prompt design

```python
prompt = f"""
You are an industrial maintenance assistant.

A machine monitoring system has detected a high risk of failure with the following sensor readings:
- Air temperature: {failure_case['Air temperature [K]']} K
- Process temperature: {failure_case['Process temperature [K]']} K
- Rotational speed: {failure_case['Rotational speed [rpm]']} rpm
- Torque: {failure_case['Torque [Nm]']} Nm
- Tool wear: {failure_case['Tool wear [min]']} minutes
- Failure probability: {failure_case['Predicted failure probability']}

Explain the likely cause of failure and suggest preventive maintenance actions.
"""
print(prompt)
```

Simulated GenAI response

```python
generated_response = """
The high probability of failure is likely due to excessive tool wear combined with elevated rotational speed and torque.
These conditions increase mechanical stress and thermal load on the system.

Recommended actions:
- Schedule immediate tool replacement
- Reduce operating speed temporarily
- Inspect cooling and lubrication systems
- Perform preventive maintenance before continued operation
"""
print(generated_response)
```

### Generative AI Observations

- Generative AI can translate numerical model outputs into human-readable explanations
- Prompt design plays a critical role in response quality
- GenAI enhances interpretability rather than replacing predictive models
- In industry, such systems are used as decision-support tools
